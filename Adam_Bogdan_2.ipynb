### Wybory 2023

```python
# Zostawiam tylko niezbędne kolumny
wybory_2023 = wybory_2023.filter(items = ["TERYT Gminy", "Gmina", "Powiat", "Województwo",
                                          "Liczba wyborców uprawnionych do głosowania", "Liczba wyborców, którym wydano karty do głosowania"
                                          ,"Liczba głosów nieważnych","Liczba głosów ważnych oddanych łącznie na wszystkie listy kandydatów"
                                          ,"Liczba wyborców głosujących na podstawie zaświadczenia o prawie do głosowania"])
# jednolite nazwy kolumn dla różnych csv
wybory_2023.rename(columns={
    'TERYT Gminy': 'TERYT',
    'Liczba wyborców uprawnionych do głosowania': 'uprawnionych_2023',
    'Liczba wyborców, którym wydano karty do głosowania': 'wydanych_2023',
    'Liczba głosów nieważnych':'Głosy nieważne_2023','Liczba głosów ważnych oddanych łącznie na wszystkie listy kandydatów':'Głosy ważne_2023',
    'Liczba wyborców głosujących na podstawie zaświadczenia o prawie do głosowania':'Zaswiadczenie_2023'
}, inplace=True)
# usuwam zagraniczne
wybory_2023 = wybory_2023[~wybory_2023["TERYT"].isna()]
#jednolity format "int"
wybory_2023['TERYT'] = pd.to_numeric(wybory_2023['TERYT'], downcast='integer')
wybory_2023
```

### Łączenie frame'ów

```python
#Podejmujemy analogiczne kroki jak przy poprzednich plikach dotyczących wyborów. Jednak ze względu na brak kolumny dotyczącej powiatu, plik wybory_2015 zostanie polączony z pozostałymi za pomocą kolumny TERYT.

wybory_razem = wybory_2023.join(wybory_2019.set_index('TERYT'), on='TERYT', validate='1:1')
wybory_razem = wybory_razem.join(wybory_2015.set_index('TERYT'), on='TERYT', validate='1:1')

wybory_razem
```

# Zadanie

```python
def election_statistics(stat, year=[2015, 2019, 2023], jst='Powiat', jst_number=10, sort_asc=True, sort_year=-1):
    """
    Funkcja election_statistics stwarza ramkę danych w oparciu o wybrane parametry:
    1)stat - statystyka którą chcemy obejrzeć, może przyjmować jeden z 3 argumentów:‘frekwencja’,'głosy nieważne','zaświadczenia'
    2)year - rok którego analizy chcemy dokonać. Domyślnie przyjmuje on wartości [2015,2019,2023]
    3)jst - określa on czy analizy chcemy dokonać na poziomie gminy, powiatu, czy województwa. W przypadku gdy wybrana zostaje gmina,
            ramka zwróci także nazwę powiatu oraz województwa w którym leżą dane gminy. Jeśli wybrany zostanie powiat, dodatkowo
            wyświetlona zostanie nazwa województwa. Domyślnie przyjmuje on wartość 'Powiat'
    4)jst_number - określa ilość jednostek terytorialnych dla których chemy dokonać analizy. Domyślnie przyjmuje wartość 10.
    5)sort - określa czy badane jednostki mają być uszeregowane rosnąco, czy malejąco. Domyślnie uszeregowane są rosnąco.
    6)sort_year - określa numer indeksu z listy year. Domyślnie przyjmuje on ostatni element listy
    
    
    Statystyki funkcji:
    
    1)Frekwencja ogółem (liczba kart wyjętych z urn/ liczba osób uprawnionych do głosowania)(‘frekwencja’)
    2)Udział głosów nieważnych (chodzi o nieważne głosy, a nie karty) w całkowitej liczbie oddanych głosów(‘głosy nieważne’)
    3)Udział wyborców głosujących na podstawie zaświadczenia o prawie do głosowaniaw całkowitej liczbie oddanych głosów(‘zaświadczenia’)
    
    Funkcja election_statistics określa jak kształtowała się wybrana statystyka dla x jednostek terytorialnych określonych zmiennymi
    jst oraz jst_number w wybranym roku (jeśli lista year ma więcej niż jeden element funkcja określa te statystyki dla roku zaindeksowanego zmienną sort_year,
    oraz porównuje wartości z wybranych jednostakch z pozostałych lat z listy year). Zmienna sort określa czy funkcja ma pokazać x jednostek
    o najmniejszej lub największej wartości wybranej statystyki.
    
    W dataframie użytym w funckji, wiersze przedstawiają poszczególne jednostki terytorialne, natomiast w kolumnach określone zostały wyniki wyborów 
    w poszczególnych latach.
    """

    wybory=wybory_razem
    allowed_columns = ["TERYT", "Gmina", "Powiat", "Województwo"]
    statistic = []
    for y in year:
        if stat == 'frekwencja':
            statistic.append(f"frekwencja_{y}")
            wybory[f'frekwencja_{y}'] = wybory[f'wydanych_{y}'] / wybory[f'uprawnionych_{y}'] * 100
        elif stat == 'głosy nieważne':
            statistic.append(f"głosy nieważne_{y}")
            wybory[f'głosy nieważne_{y}'] = wybory[f'Głosy nieważne_{y}'] / (wybory[f'Głosy nieważne_{y}'] + wybory[f'Głosy ważne_{y}']) * 100

        elif stat == 'zaświadczenia':
            statistic.append(f"zaświadczenia_{y}")
            wybory[f'zaświadczenia_{y}'] = wybory[f'Zaswiadczenie_{y}'] / (wybory[f'Głosy nieważne_{y}'] + wybory[f'Głosy ważne_{y}']) * 100


    wybory = wybory.filter(items=allowed_columns+statistic)
    wybory = wybory.groupby(jst)[statistic].mean().round(decimals=2)
    
    

    year_sort = year[sort_year]
    if stat == 'frekwencja':
        wybory = wybory.sort_values(by=f'frekwencja_{year_sort}', ascending=sort_asc)
    elif stat == 'głosy nieważne':
        wybory = wybory.sort_values(by=f'głosy nieważne_{year_sort}', ascending=sort_asc)
    elif stat == 'zaświadczenia':
        wybory = wybory.sort_values(by=f'zaświadczenia_{year_sort}', ascending=sort_asc)
    

    return wybory.head(jst_number)


```

```python
# Przykład użycia funkcji
result_df = election_statistics('frekwencja', jst='Województwo', jst_number=10, sort_asc=False)
result_df
```
